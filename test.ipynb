```python
import pandas as pd
df = pd.read_csv("data/nhl_data_plus.csv")
df = df[df["season"] != 2005] # Error year
df_2025 = df[df['season'] == 2025]
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from helper import predict_previous_season

X = df_2025[[
    "shots",
    "power_play_goals",
    "faceoff_win_pct",
    "hits",
    "blocked_shots",
    "giveaways",
    "takeaways"
]].fillna(0)
y = df_2025["won"]
X = sm.add_constant(X)

model = sm.Logit(y, X).fit()
predict_previous_season(df, model)
```

```python
probit_model = sm.Probit(y, X).fit()
predict_previous_season(df, probit_model)
```

```python
X = df_2025[[
    "shots",
    "power_play_goals",
    "faceoff_win_pct",
    "hits",
    "blocked_shots",
    "giveaways",
    "takeaways"
]].fillna(0)
y = df_2025["goals_for"].astype(int)

X = sm.add_constant(X)
```

```python
import matplotlib.pyplot as plt 

goals = df_2025["goals_for"].astype(int)
bins = np.arange(0, goals.max() + 2) - 0.5  # center

plt.hist(goals, bins=bins)

plt.xlabel("Goals scored in a game")
plt.ylabel("Number of games")
plt.title("Distribution of Goals For (2025 season)")

plt.xticks(range(0, goals.max() + 1))
plt.show()

print("Mean: ",np.mean(df_2025["goals_for"]),"SD: ", np.std(df_2025["goals_for"]))
```

```python
poisson_model = sm.GLM(
    y,
    X,
    family=sm.families.Poisson(link=sm.families.links.Log())
).fit()

print(poisson_model.summary())
```

```python
nb_model = sm.GLM(
    y,
    X,
    family=sm.families.NegativeBinomial(link=sm.families.links.Log())
).fit()

print(nb_model.summary())
```
