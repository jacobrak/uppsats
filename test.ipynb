```python
import pandas as pd
df = pd.read_csv("data/nhl_data_plus.csv")
df_2025 = df[df['season'] == 2025]
```

```python
df
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score

X = df_2025[[
    "shots",
    "power_play_goals",
    "faceoff_win_pct",
    "hits",
    "blocked_shots",
    "giveaways",
    "takeaways"
]].fillna(0)
y = df_2025["won"]

def predict_previous_season(df, model):
    average_r2_score = []
    for year in range(2004, 2025 + 1):
        df_year = df[df['season'] == year]
        X = df_year[[
            "shots",
            "power_play_goals",
            "faceoff_win_pct",
            "hits",
            "blocked_shots",
            "giveaways",
            "takeaways"
        ]].fillna(0)
        y = df_year["won"]
        prediction = model.predict(X)
        prediction = prediction.round().astype(int)
        score = accuracy_score(y, prediction)
        print(f"Accuracy score for season {year}: {score}")
        average_r2_score.append(score)
    print(f"Average accuracy score: {np.mean(average_r2_score)} over {len(average_r2_score) - 1} Test Seasons")

model = LinearRegression()
model.fit(X, y)
predict_previous_season(df, model)
```

```python
model = LogisticRegression()
model.fit(X, y)
predict_previous_season(df, model)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)

probit_model = sm.Probit(y, X)
probit_results = probit_model.fit()

print(probit_results.summary())
```
